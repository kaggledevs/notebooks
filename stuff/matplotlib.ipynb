## Simple plots and their settings

We will draw `sin` and `cos` functions on one plot.

Get the data:

```python
# 256 points from -π to +π included
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)

# sins, coses for each x in X
C, S = np.cos(X), np.sin(X)
```


Draw the plot:

```python
# Create a figure 8x6 inches, 80 dpi
plt.figure(figsize=(10,6), dpi=80)

# Create a new subplot from a grid 1x1
plt.subplot(1, 1, 1)

# Plot the cosine with blue continuos line width 1
plt.plot(X, C, color="blue", linewidth=2.5, linestyle='-')
# Plot the sine with green continuos line width 1
plt.plot(X, S, color="red", linewidth=2.5, linestyle='-')


# Set x limits and ticks
plt.xlim(-4.0, 4.0)
plt.xticks(np.linspace(-4, 4, 9, endpoint=True))

# Set y limits and ticks
plt.ylim(-1, 1)
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))
```

### Setting Limits and Ticks

We've already done that for the plot above.

But current limits are not enough. Make them larger.

After we make them larger, we'll replace x-ticks with strings (for example, 3.14 -> "π"). We use **LaTeX** to allow for nice rendering of the label.

```python
plt.figure(figsize=(10,6), dpi=80)
plt.subplot(1, 1, 1)

plt.plot(X, C, color="blue", linewidth=2.5, linestyle='-')
plt.plot(X, S, color="red", linewidth=2.5, linestyle='-')


# Set limits
plt.xlim(X.min() * 1.1, X.max() * 1.1)
plt.ylim(C.min() * 1.1, C.max() * 1.1)

# Set correct ticks and corrsepoding labels
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

plt.yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$'])
```

### Adding Legend

* `plt.legend()` - to add a legend somewhere on the canvas. Requires adding the keyword argument label (that will be used in the legend box) to the plot commands.
* `plt.annotate()` - to create an annotation: a piece of text referring to a data point (optionally, with an arrow).

```python
plt.figure(figsize=(10,6), dpi=80)
plt.subplot(1, 1, 1)

plt.plot(X, C, color="blue", linewidth=2.5, linestyle='-', label="cosine")
plt.plot(X, S, color="red", linewidth=2.5, linestyle='-', label="sine")

plt.xlim(X.min() * 1.1, X.max() * 1.1)
plt.ylim(C.min() * 1.1, C.max() * 1.1)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$'])

# Add legend
plt.legend(loc='upper left')
```
